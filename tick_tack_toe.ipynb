```python
def print_board(board_positions):
    max_input_width = len(str(max([max(i) for i in board_positions])))
    cell_width = max_input_width + 2 # "2" is padding for to make the cell contents more visable
    cols = max(map(len, board_positions))
    #set top border as first line in string
    rtn_str = ['+' + ('-'*(cell_width) + '+')*cols, '\n']

    for row in board_positions:
        # add data line in string w/ padding 
        for value in row:
            rtn_str += f'| {value} '

        # add string line for bottom of cell   
        rtn_str+=['|\n+',('-'*(cell_width) +'+' )*cols,'\n']

    rtn_str = ''.join(rtn_str)
    print(rtn_str)
    return None
```
